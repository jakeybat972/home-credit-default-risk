## Decision Thresholding Using Expected Value

Rather than selecting a probability threshold arbitrarily, we evaluate an approval policy using a credit-risk style expected value framework:

- PD: model-predicted probability of default
- EAD: exposure at default, approximated by `AMT_CREDIT`
- LGD: loss given default (assumed)
- margin: expected profit on a performing loan (assumed)

Expected Value of approving a loan:

EV(approve) = (1 - PD) * margin * EAD - PD * LGD * EAD

Approve loans when EV(approve) >= 0.

Assumptions are documented and can be stress-tested.
